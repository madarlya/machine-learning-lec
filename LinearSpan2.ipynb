## Linear span 

```python
import numpy as np
```

```python
# does {v1,v2,v3} span R ? 

v1 = np.array([1,0,0])
v2 = np.array([0,1,0])
v3 = np.array([0,0,1])    #  dependant : v3 = v1 + v2
A = np.column_stack([v1,v2,v3])
print("rank: ", np.linalg.matrix_rank(A))
```

```python
# Orthonormal basis for span of columns (QR)

v4 = np.array([0,0,1])
b = np.column_stack([v1,v2,v4])  # independant
Q,R = np.linalg.qr(b)
print("Orthonormal basis for span : ", Q)
```

```python
# SVD approach : basis vectors for span

U,S,Vt = np.linalg.svd(A, full_matrices=False)
r = np.sum(S > 1e10)
print("Span basis (U cols) : ", U[:,:r])
```

```python
# is b in the span A's columns ?

B = np.array([2,3,1])
x, res , rank, sv = np.linalg.lstsq(A,B, rcond=None)
B_proj = A @ x
print("b in span ? ", np.allclose(B,B_proj))
```
