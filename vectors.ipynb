A vector is a mathematical object that encodes a mathematical object that encodes a length & direction 

more formally they are elements of a vector space, a collection of object that is closed under an addition rule
& rule for multiplication by scalars 

a vector is often represented as 1-dimendional (1-D) array of numbers reffered to as componant ant its displayed either in column form or row form

```python
import numpy as np 
```

```python
v = np.array([1,-1,3])
w = np.array([2,5,2])
```

```python
print(v+w)
```

```python
print(v-w)
```

### scalar multiplication

```python
print(3*v)
```

### find length of v

```python
print(np.linalg.norm(v))
```

### dot product

```python
print(np.dot(v,w))
```
