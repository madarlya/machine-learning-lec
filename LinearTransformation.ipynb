## Linear Transformation 

```python
import numpy as np
import scipy 
```

```python
# define a linear transformation as matrix A

A = np.array([[2,0],
              [0,3],
              [1,-1]])

u = np.array([1,2])
v = np.array([3,-1])
c = 4.0
```

```python
# apply transformation 

Tu = A @ u
Tv = A @ v
print(Tu , Tv)
```

```python
# verify axiom 1: T(u=v) == T(u)+T(v)

print(np.allclose(A @ (c*v), c * Tv))
```

### matrix representation

```python
def T(v):
    x,y = v
    return np.array([x+y,2*x,x-y])
```

```python
# build matrix by applying T to each standard basis vector 

n = 2 
cols = [T(np.eye(n)[:,i])for i in range (n)]
A = np.column_stack(cols)
print(A)
```

```python
# verify A @ v == T(v) for any v

v = np.array([3,-2])
print(np.allclose(A @ v, T(v)))
```

```python
# change of basis : same T, difrent coordinate system

B = np.array([[1,1],[0,1]],dtype = float)  # new basis
A_in_B = np.linalg.solve(B,A[:2]@ B)
```
