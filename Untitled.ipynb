```python
import numpy as np
```

```python
# vector addition 
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])

print(v1 + v2)
```

```python
# vector substraction 
print(v1-v2)
```

```python
# scalar multiplication

print(3  * v1)
```

```python
# dot product

print(np.dot(v1,v2))
```

```python
# element wise multiplication 
print(v1*v2)
```
