vectors in Rn + vector algebra 

```python
import numpy as np
```

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
```

#### addition & substraction


```python
add = v1 + v2
sub = v1 - v2
print(sub)
print(add)
```

#### Magnitude

```python
mag = np.linalg.norm(v1)
print(mag)
```

#### linear combination of vector

```python
c1, c2 = 2,-1
result = c1 * v1 + c2 * v2
print(result)
```

#### linear independance check

```python
A = np.array([[1,2],
              [2,4]])
rank = np.linalg.matrix_rank(A)
if rank < A.shape[1]:
    print("dependant")
else:
    print("Independant")
```

#### Orthonormal Vectors

```python
v4 = np.array([1,0])
v5 = np.array([0,1])

dot = np.dot(v4,v5)
norm_v4 = np.linalg.norm(v4)

print("Dot",dot)
print("Norm",norm_v4)
```

```python
x = np.array([[1,1],
              [1,2],
              [1,3]])
y = np.array([2,3,4])

theta = np.linalg.inv(x.T.dot(x)) .dot(x.T) .dot(y)
print(theta)
```
